```python
# 'http://stats.nba.com/league/player/#!/?Season=2015-16&SeasonType=Regular%20Season&TeamID=1610612757')
source = open("blzr_2015_playerstats.csv").read()
soup = bs4.BeautifulSoup(source)
```

```python
table = soup.find('table', {'class': 'table'})
```

```python
# Create four variables to score the scraped data in
# class = ng-binding 

player =[]
team = []
age =[]
games_played = []
losses = []
wins = []
minutes = []
field_goal_percentage = []
three_points_made = []
three_points_attempted = []
three_point_percentage = []
free_throw_percentage = []
double_doubles = []
triple_double = []
plus_minus =[]

# class = ng-binding.stats-popup-link

field_goals_made = []
field_goals_attempted = []
free_throw_made = []
free_throw_attempted = []
offensive_rebound = []
defensive_rebound = []
rebounds = []
assists = []
turnovers = []
steals = []
blocks = []
personal_fouls = []
points = []

for row in table.findAll('tr')[1:]:
    col = row.findAll('td') 
    player.append(str(col[0].text))
    team.append(col[1].text)
    age.append(col[2].text)
    games_played.append(col[3].text)
    wins.append(col[4].text)
    losses.append(col[5].text)
    minutes.append(col[6].text)
    field_goals_attempted.append(str((col[7].text)).replace('\n\n\n\nVideo\nShotchart\n\n\n\n\n\n', ''))
    field_goals_made.append(str((col[8].text)).replace('\n\n\n\nVideo\nShotchart\n\n\n\n\n\n', ''))
    field_goal_percentage.append(col[9].text)
    three_points_made.append(str((col[10].text)).replace('\n\n\n\nVideo\nShotchart\n\n\n\n\n\n', ''))
    three_points_attempted.append(str((col[11].text)).replace('\n\n\n\nVideo\nShotchart\n\n\n\n\n\n', ''))
    three_point_percentage.append(col[12].text)
    free_throw_made.append(str((col[13].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    free_throw_attempted.append(str((col[14].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    free_throw_percentage.append(col[15].text)
    offensive_rebound.append(str((col[16].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    defensive_rebound.append(str((col[17].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    rebounds.append(str((col[18].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    assists.append(str((col[19].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    steals.append(str((col[20].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    turnovers.append(str((col[21].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    blocks.append(str((col[22].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    personal_fouls.append(str((col[23].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    double_doubles.append(col[24].text)
    triple_double.append(col[25].text)
    points.append(str((col[26].text)).replace('\n\n\n\nVideo\n\n\n\n\n\n\n', ''))
    plus_minus.append(col[27].text)
```
